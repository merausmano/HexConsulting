### Categorical Variable Handling ###

Earlier we came across the feature *'view'* which should be handled as a categorical variable but has not been taken up. The feature *'sqft_basement'* also requires treatment as a categorical variable. However, the treatment for this is more complex.

The reason for the complexity is that this feature is operating in two different modes. In the first, a distinction needs to be made between possessing or not possessing a basement. In the second, if there is a basement, the area of the basement acts like a continuous variable on its impact on *'price'* . Contrast the plot of the feature *'view'* with *'price'* (in the second batch of plots) against the plot of the feature *'sqft_basement'* with *'price'* (in the third batch of plots) .

The procedure adopted is as follows:
1. Identify the rows where there is a basement and create a new 'sqft_basement' variable for these rows only
1. The other rows become NaN (not available)
1. Plot to check that there is a linear relationship between the modified feature and the dependent variable
1. Fill the NaNs with zeros (or other suitable values)
1. Create a new categorical variable that is composed of zeros and ones
1. Zeros are for the rows where there is no basement and one are those with a basement


```python
base_area = dh.sqft_basement[dh.sqft_basement>0]    
dh['sqft_basement'] = base_area

# The two lines above replace all the zero values with NaNs; see the next cell
# This also separates the two effects (zero and non-zero); see the plot below
```

```python
# Introduction of NaNs separates the two effects (zero and non-zero)
# The plot below shows the effect of the non-zeros 

plt.scatter((dh.sqft_basement), dh.price)
(dh.sqft_basement).corr(dh.price), dh.price.corr(np.log(dh.sqft_basement))
```

### Regressions

```python
'''
Not all of the independent variables of the dataset will be used in the 
regression models that follow. The features that would be used are being
separated from the other variables and collected together in 'X'.

What type of object is X?

After assembing the relevant columns in X, examine the contents of X.
'''

Xcols = [2, 3, 4, 8, 10, 11, 12, 16, 18]
Xcolnames = []
for i in Xcols:
    Xcolnames.append(dh.columns[i])
    X = dh[Xcolnames]

X.head()
```

```python
'''
Examining X reveals the presence of NaNs. These need to be replaced.
We have NaNs in only one variable, but we'll check this later.

The NaNs are replaced by zeros and, then, a new categorical variable is created

This variable is called 'bassement' and is added as a feature to X.
It contains 0 or 1, instead of True or False.

We use numerical substitutes for Boolean values as numerical values are
more easily handled in regression algorithms.
'''

X.sqft_basement = X.sqft_basement.fillna(0)
X['basement'] = (X.sqft_basement > 0) * 1
X.head()
```

#### Traditional Regression ####

The first model (using **statsmodels.formula.api**) will take a traditional approach, with traditional output. It requires:
1. The dependent variable to be in the same dataset as the independent variables
2. The independent variables to be specified by name
3. It will automatically add a constant - in other statsmodels libraries, you may have to add the constant yourself

```python
XwithDV = pd.concat([dh.price, X], axis = 1)

# The dependent variable is combined with X into a new object.
# This is done by the 'concat' function, which is part of Pandas
# axis = 1 is needed for extending horizontally

# What type is the new object?

# The command above adds back price to the relevant set of independent features
# Below we check for missing values in the relevant object (XwithDV)
# It sums across all columns after checking each row in a column.

sum(XwithDV.isnull().sum() != 0)
```

```python
'''
Specification of the regression model equation - without an intercept

Note the .fit() at the end!
It is responsible for calculating the actual values in the model 
The .ols portion only decides which equations need to be calculated

'''


Model1 = smfa.ols('price ~ bedrooms + bathrooms	+ sqft_living +	view + grade + sqft_above +	sqft_basement + lat + sqft_living15 + basement', data=XwithDV).fit()
```

You can try your own models by adjusting the above **smfa.ols** command suitably

```python
# Having created and calculated the model in the previous cell,
# the results are printed out:

print (Model1.summary())

# Examine the results. What conclusions can you derive from the results shown?
```

##### Modelling 'view' properly

The scatter plots before regression showed that *'view'* is not a continuous variable, so the 
best treatment (in this type of regreression algorithm) is to treat it as a categorical variable.
We need to process the variable further to use it as such. The procedure is described below

* The categorical variable "View" has 5 states
* Using a special function (or otherwise), generate new variables for each state of view
* Any one of the 5 variables needs to be discarded to avoid multicollinearity errors.
* The original variable also needs to be discarded.
* The first one is absorbed in the intercept, leaving 4 variables
* The variables are named below and then added to the existing DataFrame
* A new DataFrame is created for a new regression


```python
'''
get_dummies is the special function used to create and fill the new dummy
variables with 0 and 1 in the relevant places.

For example, the last variable created will have a 1 when view = 5, otherwise 0
This logic is implemented in the other newly created variables.
It is called: One-Hot-Encoding (OHE)

'a' stores the newly generated variables (columns)
'''


a = pd.get_dummies(dh.view, drop_first=True)
a.columns = ['v1', 'v2', 'v3', 'v4']
```

```python
XwithDV2 = pd.concat([XwithDV, a], axis=1)

XwithDV2.head()
```

```python
# This is the specification of the new model, with additional dummy variables
# Remember, you can create your own models, too.

Model1V = smfa.ols('price ~ bedrooms + bathrooms	+ sqft_living +	grade + sqft_above +	sqft_basement + lat + sqft_living15 + basement + v1 + v2 + v3 +v4', data=XwithDV2).fit()

print(Model1V.summary())
```

#### Machine Learning Regression Model
In the next regression model, using **sklearn.linear_model**, the independent variables need to be separated from the
dependent variable. However, it is also useful to separate out a training sample (**training data**) and a testing sample (**test data**).

Below, 80% of the data is used for training. The model generated from this training is tested on the remaining 20%. The
various performance measures refer to the performance of the model with the test data.

A specific random state is chosen to ensure repeatability of results.

*Next steps in the machine learning procedure*

* The dependent variable is re-named
* Training and testing samples are generated
* Training data is uses to test the remaining data
* Note that the remaining data has not been used in deriving the model
* This is a good precaution against overfitting.

```python
y = dh.price


# Splitting the dataset into train and test portions are done by
# train_test_split - currently found in sklearn.Linear_Regression

X_train, X_test, y_train, y_test = train_test_split(X, y ,test_size=0.2, random_state = 21)
```

* Now, the linear regression model is to be invoked and labelled. 
* Invocation loads the relevant data structures and model equations.
* This model is then 'fitted' with the data. 
* Fitting calculates the coefficients of the various model equations.
* Note that these coefficients are based on the training dataset.

```python
Model2 = LinearRegression()
Model2.fit(X_train, y_train)
```

The coefficients of the trained (or fitted) model are derived from fitting the training data, as above.

These coefficients are then applied to the test data to make predictions.

```python
Coeffs = list(Model2.coef_)           # extract the coefficients into a list
Coeffs.insert(0, Model2.intercept_)   # insert intercept at the start of list

m = (Model1.summary2().tables[1])     # locate the coefficient names of Model 1

Coeff_names = m.index.to_list()       # extract the names into a list


# A DataFrame object is created that assembles the list of coefficient names
# and the coefficients

pd.DataFrame({'Coeff_names': Coeff_names, 'Coeff_values': Coeffs})
```

##### Fitting and Prediction #####

The coefficients from the **fitted** model are now applied to the test data to make **predictions**.

```python
ypred2 = Model2.predict(X_test)

y_train_pred2 = Model2.predict(X_train)
```

```python
Model2.score(X_test,y_test), Model2.score(X_train, y_train)
```

The score above is similar to the measure known as "R-squared"

* If the score based on test data is lower than the score based on the training data, it indicates overfiiting.
* Conversely, if the score based on test data is higher than the score of the training data, then it is an indication of underfitting.
* There is slight underfitting, as seen from the above results.

The measure below extract mean-squared-error and root-mean-square-error


```python
print("MSE:", np.mean((ypred2 - y_test) ** 2))
print("RMSE:", np.math.sqrt(np.mean((ypred2 - y_test) ** 2)))
```

### Other accuracy measures

Now, a different way of measuring the accuracy of the predicted results is proposed. These are very close to MAPE or mean absolute percentage error

* Using the predictions from the last XGBoost model, instead of calculating errors based on log values, the actual *absolute* difference (called **Difference**) and the actual *error-measuring* ratio is calculated.
* Intermediate variables created in this calculation are: *actual_ratio, inverse_ratio and error_ratio*
* A new DataFrame object called **compare_actual** is then created in the cell after that
* The variable called **Ratio** in the last column is the same as the variable *actual_ratio* in the calculations
* After displaying some aspects of the new DataFrame, the average measure and of **error_ratio** is calculated in the cell after that.
* It also shows the average of *actual_ratio* and the stanard deviation of *error_ratio*


```python
actual_y_test = np.exp(yml_test)                  # obtain actual values of y

actual_predicted = np.exp(predictions_xg5)        # actual predicted values of y

diff = abs(actual_y_test - actual_predicted)      # absolute difference

actual_ratio = actual_predicted / actual_y_test   # ratio of the two

inverse_ratio = 1/actual_ratio                    # obtain inverse of the ratios

dj = pd.concat([actual_ratio, inverse_ratio], axis = 1) # assemble the ratios

error_ratio = dj.apply(lambda p: p.max(), axis =1) # choose maximum of the two

# This ensures that the ratio is always greater than 1.0
```

```python
# Create a new DataFrame with actual values, not logged values
# Name of the DataFrame is 'compare_actual'
# The DataFrame object is created with 3 columns: Test Data, Predicted_Price and Difference

compare_actual = pd.DataFrame({'Test Data': actual_y_test, 'Predicted_Price' : actual_predicted, 'Difference' : diff})


# Now 2 more columns are created for this DataFrame
compare_actual['Diff_%'] = diff/actual_y_test * 100
compare_actual['Ratio'] = actual_ratio

# First 3 columns are set as integers
compare_actual.iloc[:, :3] = compare_actual.iloc[:, :3].astype(int)
compare_actual.head(6)
```

```python
print ("MAPE :", np.mean(compare_actual['Diff_%']), " st. dev. of MAPE :", np.std(compare_actual['Diff_%']))
print ("Actual ratio (mean) :", np.mean(compare_actual.Ratio), "Error_Ratio (mean):", np.mean(error_ratio))
```
